```python
shortP = 50
longP = 120

signaldf = pd.DataFrame(index = spydf.index)
signaldf['signal'] = 0.0

signaldf['short_mav'] = spydf['close'].rolling(window=shortP, min_periods=1, center=False).mean()
signaldf['long_mav'] = spydf['close'].rolling(window=longP, min_periods=1, center=False).mean()

signaldf['signal'][shortP:] = np.where(signaldf['short_mav'][shortP:] > signaldf['long_mav'][shortP:], 1.0, 0.0)
signaldf['positions'] = signaldf['signal'].diff()
signaldf[signaldf['positions'] == -1.0]
```

```python
fig = plt.figure()
plt1 = fig.add_subplot(111, ylabel='Price in $')

spydf['close'].plot(ax=plt1, color='r', lw=2.)

signaldf[['short_mav', 'long_mav']].plot(ax=plt1, lw=2., figsize=(12, 8))

plt1.plot(signaldf.loc[signaldf.positions == -1.0].index, signaldf.short_mav[signaldf.positions == -1.0], 'v', markersize=10, color='k')
plt1.plot(signaldf.loc[signaldf.positions == 1.0].index, signaldf.short_mav[signaldf.positions == 1.0], '^', markersize=10, color='m')

plt.show()
```
